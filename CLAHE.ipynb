```python
import cv2
import os
import zipfile
from google.colab import drive
from google.colab import files
```

```python
# Mount your Google Drive
drive.mount('/content/drive')
```

```python
# Define the paths
input_folder = '/content/drive/My Drive/Dental_opg'
output_folder = '/content/drive/My Drive/output'
```

```python
print(input_folder)
```

```python
# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Set CLAHE parameters
clip_limit = 8.0
tile_grid_size = (20, 20)

clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)

for filename in os.listdir(input_folder):
    input_path = os.path.join(input_folder, filename)
    if filename.endswith('.jpg') or filename.endswith('.png') or filename.endswith('.jpeg'):
        # Read the image
        image = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)
        clahe_image = clahe.apply(image)


        output_path = os.path.join(output_folder, filename)
        cv2.imwrite(output_path, clahe_image)

        print(f"Complete : {output_path}")
```

```python
# Zip the output folder
output_zip = '/content/drive/My Drive/output.zip'
with zipfile.ZipFile(output_zip, 'w') as zipf:
    for root, dirs, files in os.walk(output_folder):
        for file in files:
            zipf.write(os.path.join(root, file), os.path.relpath(os.path.join(root, file), output_folder))

print(f"Zipped: {output_zip}")

# Download the zip file (if running on Google Colab)
from google.colab import files
files.download(output_zip)
```
